```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing = pandas.read_csv("D:/Statheros_Data/GlobalHousePriceIndex.csv")
btc = pandas.read_csv("D:/Statheros_Data/CBBTCUSD.csv")
usd = pandas.read_csv("D:/Statheros_Data/DTWEXBGS.csv")
gold = pandas.read_csv("D:/Statheros_Data/GOLDAMGBD228NLBM.csv")
usd_long = pandas.read_csv("D:/Statheros_Data/USD_Long.csv")
gold_long = pandas.read_csv("D:/Statheros_Data/GOLD_Long.csv")
```

```python
housing_long = housing
usd = usd.rename(columns={'DTWEXBGS_NBD20150101': 'USD Indexed at 2015'})
housing = housing.rename(columns={'equally weighted': 'World Housing Indexed at 2015', 'dateq' : 'DATE'})
btc = btc.rename(columns={'CBBTCUSD_NBD20150101': 'BTC Indexed at 2015'})
gold = gold.rename(columns={'GOLDAMGBD228NLBM_NBD20150101': 'Gold Indexed at 2015'})
usd_long = usd_long.rename(columns={'DTWEXBGS': 'USD Indexed at 2006'})
gold_long = gold_long.rename(columns={'GOLDAMGBD228NLBM_NBD20060101': 'Gold Indexed at 2006'})
housing_long = housing_long.rename(columns={'equally weighted': 'World Housing Indexed at 2006', 'dateq' : 'DATE'})
```

```python
usd.drop(usd.index[:36], inplace=True)
housing["World Housing Indexed at 2015"] =  housing["World Housing Indexed at 2015"] / housing.at[60, "World Housing Indexed at 2015"] * 100
housing.drop(housing.index[:60], inplace=True)
housing_long["World Housing Indexed at 2006"] =  housing_long["World Housing Indexed at 2006"] / housing_long.at[24, "World Housing Indexed at 2006"] * 100
housing_long.drop(housing_long.index[:24], inplace=True)
housing['DATE'] = pd.to_datetime(housing['DATE'])
housing_long['DATE'] = pd.to_datetime(housing_long['DATE'])
btc['DATE'] = pd.to_datetime(btc['DATE'])
usd['DATE'] = pd.to_datetime(usd['DATE'])
gold['DATE'] = pd.to_datetime(gold['DATE'])
usd_long['DATE'] = pd.to_datetime(usd_long['DATE'])
gold_long['DATE'] = pd.to_datetime(gold_long['DATE'])
```

```python
ax = plt.gca()

housing_long.plot(kind='line',x='DATE',y='World Housing Indexed at 2006',ax=ax)
usd_long.plot(kind='line',x='DATE',y='USD Indexed at 2006', color='red', ax=ax)
# the plot gets saved to 'output.png'
plt.savefig('D:/Statheros_Data/USDvsHousing.png')
plt.show()
```

```python
ax = plt.gca()

housing_long.plot(kind='line',x='DATE',y='World Housing Indexed at 2006',ax=ax)
usd_long.plot(kind='line',x='DATE',y='USD Indexed at 2006', color='red', ax=ax)
gold_long.plot(kind='line',x='DATE',y='Gold Indexed at 2006', color='green', ax=ax)
# the plot gets saved to 'output.png'
plt.savefig('D:/Statheros_Data/GoldvsOthers.png')
plt.show()
```
